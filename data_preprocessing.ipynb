```python
import pandas as pd

# 1. 데이터 로드 (이전과 동일)
df_orders = pd.read_csv(r'data\blinkit_orders.csv')
df_feedback = pd.read_csv(r'data\blinkit_customer_feedback.csv')
df_delivery = pd.read_csv(r'data\blinkit_delivery_performance.csv')
df_items = pd.read_csv(r'data\blinkit_order_items.csv')

# 2. 데이터 병합 (주문 ID 기준)
df_order_merged = pd.merge(df_items, df_orders, on='order_id', how='left')
df_order_merged = pd.merge(df_order_merged, df_delivery[['order_id', 'delivery_time_minutes']], on='order_id', how='left')
df_order_merged = pd.merge(df_order_merged, df_feedback[['order_id', 'rating', 'feedback_category']], on='order_id', how='left')

# 3. 배송시간 계산 (실제_배송일시 - 주문일시)
# 날짜 형식으로 변환 후 계산해야 합니다.
df_order_merged['order_date'] = pd.to_datetime(df_order_merged['order_date'])
df_order_merged['actual_delivery_time'] = pd.to_datetime(df_order_merged['actual_delivery_time'])

# 두 시간의 차이를 구한 뒤 분(.total_seconds() / 60)으로 변환
df_order_merged['배송시간'] = (df_order_merged['actual_delivery_time'] - df_order_merged['order_date']).dt.total_seconds() / 60

# 4. 필요한 컬럼 선택 및 한글화 (store_id 제외, 배송시간 추가)
df_order_final = df_order_merged[[
    'order_id', 'customer_id', 'order_date', 'actual_delivery_time',
    '배송시간', 'delivery_time_minutes', 'order_total', 
    'rating', 'feedback_category', 'product_id', 'quantity', 'unit_price'
]]

df_order_final.columns = [
    '주문_ID', '고객_ID', '주문일시', '실제_배송일시', 
    '배송시간', '배송지연시간', '주문_총액', 
    '평점', '피드백_카테고리', '상품_ID', '수량', '단가'
]

# 결과 확인
print("--- 배송시간 컬럼 추가 및 병합 완료 ---")
print(df_order_final[['주문일시', '실제_배송일시', '배송시간']].head())
```

```python
import pandas as pd

# 데이터 로드
df_delivery = pd.read_csv(r'data\blinkit_delivery_performance.csv')
df_items = pd.read_csv(r'data\blinkit_order_items.csv')

# 2. 병합 (주문_ID 기준)
# 배송소요_분(delivery_time_minutes)을 가져오기 위해 병합합니다.
df_step1 = pd.merge(
    df_items, 
    df_delivery[['order_id', 'delivery_time_minutes']], 
    on='order_id', 
    how='left'
)

# 3. 컬럼명 한글로 변경
mapping = {
    'order_id': '주문_ID',
    'product_id': '상품_ID',
    'quantity': '수량',
    'unit_price': '단가',
    'delivery_time_minutes': '배송_지연시간'
}

df_step1 = df_step1.rename(columns=mapping)

# 결과 확인
print(df_step1.head())
```

```python
df_step1['금액'] = df_step1['수량']*df_step1['단가']
df_step1 = df_step1.drop(columns=['단가'])
df_step1
```

```python
import pandas as pd

# 1. 피드백 데이터 로드
df_feedback = pd.read_csv(r'data\blinkit_customer_feedback.csv')

# 2. df_step1과 병합 (주문_ID 기준)
# df_feedback의 'order_id'와 df_step1의 '주문_ID'를 매칭하여 'rating'을 가져옵니다.
df_step2 = pd.merge(
    df_step1,
    df_feedback[['order_id', 'rating']],
    left_on='주문_ID',
    right_on='order_id',
    how='left'
)

# 3. 컬럼명 정리 및 중복 ID 삭제
df_step2 = df_step2.rename(columns={'rating': '평점'}).drop(columns=['order_id'])

# 결과 확인
print(df_step2.head())
```

```python
import pandas as pd

# 1. 주문 기본 정보 로드
df_orders = pd.read_csv(r'data\blinkit_orders.csv')

# 2. 날짜/시간 데이터 변환 및 계산
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_orders['actual_delivery_time'] = pd.to_datetime(df_orders['actual_delivery_time'])

# 배송소요시간 계산 (실제 배송 - 주문 시간) 분 단위
df_orders['배송소요시간'] = (df_orders['actual_delivery_time'] - df_orders['order_date']).dt.total_seconds() / 60

# 주문날짜 추출 (시간 정보 제외)
df_orders['주문날짜'] = df_orders['order_date'].dt.date

# 3. df_step1과 병합을 위한 컬럼 정리
df_orders_subset = df_orders[['order_id', 'customer_id', '배송소요시간', '주문날짜']]
df_orders_subset = df_orders_subset.rename(columns={'order_id': '주문_ID', 'customer_id': '고객_ID'})

# 4. 최종 병합
df_step3 = pd.merge(df_step2, df_orders_subset, on='주문_ID', how='left')

# 컬럼 순서 보기 좋게 정리
df_step3 = df_step3[['주문_ID', '고객_ID', '상품_ID', '주문날짜', '금액', '배송_지연시간', '배송소요시간', '수량','평점']]

# 결과 확인
print(df_step3.head())
```

```python
import pandas as pd

# 1. 상품 정보 로드 및 병합
df_products = pd.read_csv(r'data\blinkit_products.csv')
# 필요한 상품 컬럼 선택 및 한글화
df_prod_subset = df_products[['product_id', 'product_name', 'category', 'brand', 'shelf_life_days']].rename(columns={
    'product_id': '상품_ID',
    'product_name': '상품명',
    'category': '카테고리',
    'brand': '브랜드',
    'shelf_life_days': '유통기한_일수'
})
df_step4 = pd.merge(df_step3, df_prod_subset, on='상품_ID', how='left')

# 2. 고객 정보 로드 및 병합
df_customers = pd.read_csv(r'data\blinkit_customers.csv')
# 필요한 고객 컬럼 선택 (customer_id -> 고객_ID로 매칭)
df_cust_subset = df_customers[['customer_id', 'area', 'customer_segment']].rename(columns={
    'customer_id': '고객_ID',
    'area': '지역',
    'customer_segment': '회원타입'
})
df_step5 = pd.merge(df_step4, df_cust_subset, on='고객_ID', how='left')

# 3. 컬럼 순서 정리
df_step5 = df_step5[['주문_ID', '고객_ID', '주문날짜', '상품_ID', '상품명', '카테고리', '브랜드', '유통기한_일수', '지역', '회원타입', '금액', '배송_지연시간', '배송소요시간', '수량','평점']]

# 결과 확인
print(df_step5.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# 1. 데이터 로드
df = pd.read_csv('blinkit_final_data.csv')

# 2. Label Encoding (컬럼 개수를 유지하며 숫자로 변환)
# 글자로 된 컬럼들만 선택해서 숫자로 바꿉니다.
cols_to_encode = ['카테고리', '브랜드', '상품명', '회원타입', '지역']
df_label = df.copy()

le = LabelEncoder()
for col in cols_to_encode:
    # 각 컬럼의 고유한 글자들을 0, 1, 2... 숫자로 매핑
    df_label[col] = le.fit_transform(df_label[col].astype(str))

# 3. 전체 상관관계 계산
# 이제 모든 데이터가 숫자이므로 전체 컬럼 간의 상관관계를 한꺼번에 구할 수 있습니다.
# 날짜 컬럼을 제외하고 상관관계 계산
# numeric_only=True 옵션을 주면 숫자 형태의 컬럼만 자동으로 골라냅니다.
corr_matrix = df_label.drop(columns=['주문날짜']).corr()

# 4. 히트맵 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap (Label Encoded)', fontsize=15)
plt.show()

# 5. 수량(적정 재고)과 매출(금액)에 대한 상관계수만 확인
print("--- 수량 및 금액 기준 상관계수 ---")
print(corr_matrix[['수량', '금액']].sort_values(by='수량', ascending=False))
```

```python
import pandas as pd

# 1. 데이터 로드 및 날짜 처리
df = pd.read_csv('blinkit_final_data.csv')
df['주문날짜'] = pd.to_datetime(df['주문날짜'])

# 2. 상품명별 일일 집계 (Daily Aggregation)
# '주문 증가량'과 '평점 변화'를 보기 위해 일별로 데이터를 묶습니다.
df_daily = df.groupby(['주문날짜', '상품명']).agg({
    '수량': 'sum',           # 일일 총 판매량 (Target)
    '금액': 'sum',           # 일일 총 매출
    '평점': 'mean',          # 일일 평균 평점
    '배송소요시간': 'mean',   # 일일 평균 배송시간
    '카테고리': 'first',     # 범주형 정보 유지
    '브랜드': 'first'
}).reset_index().sort_values(['상품명', '주문날짜'])

# 3. [핵심] 시계열 피처 생성 (주문 증가량 및 평점 변화)
# 각 상품별로 이전 날짜의 데이터를 가져옵니다.
for lag in [1, 7]: # 1일 전, 7일 전(전주) 데이터
    # 판매 증가량 계산을 위한 과거 판매량
    df_daily[f'수량_lag_{lag}'] = df_daily.groupby('상품명')['수량'].shift(lag)
    # 평점 변화를 보기 위한 과거 평점
    df_daily[f'평점_lag_{lag}'] = df_daily.groupby('상품명')['평점'].shift(lag)

# 4. 증가율/변화율 계산
df_daily['주문_증가량'] = df_daily['수량'] - df_daily['수량_lag_1']
df_daily['평점_변화'] = df_daily['평점'] - df_daily['평점_lag_1']

# 결측치 제거 (첫 날짜 등 데이터가 없는 경우)
df_daily = df_daily.dropna()

print("--- 모델링용 시계열 데이터셋 완성 ---")
print(df_daily[['주문날짜', '상품명', '수량', '주문_증가량', '평점_변화']].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. 데이터 로드 및 시계열 변환
df = pd.read_csv('blinkit_final_data.csv')
df['주문날짜'] = pd.to_datetime(df['주문날짜'])

# 2. 상품명별 일일 집계 (모델링 단위: 상품명 + 날짜)
# 수요 예측의 밀도를 높이기 위해 상품명으로 그룹화합니다.
df_daily = df.groupby(['주문날짜', '상품명']).agg({
    '수량': 'sum',           # Target: 적정 재고 예측용
    '금액': 'sum',           # Target: 매출 예측용
    '평점': 'mean',
    '배송소요시간': 'mean',
    '카테고리': 'first',
    '브랜드': 'first'
}).reset_index().sort_values(['상품명', '주문날짜'])

# 3. 피처 엔지니어링 (중요 변수 생성)
# (1) 날짜 관련 변수
df_daily['month'] = df_daily['주문날짜'].dt.month
df_daily['day_of_week'] = df_daily['주문날짜'].dt.dayofweek
df_daily['is_weekend'] = df_daily['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# (2) 시계열 Lag 변수 (과거 데이터)
df_daily['수량_lag_1'] = df_daily.groupby('상품명')['수량'].shift(1) # 어제 판매량
df_daily['수량_lag_7'] = df_daily.groupby('상품명')['수량'].shift(7) # 지난주 오늘 판매량
df_daily['평점_lag_1'] = df_daily.groupby('상품명')['평점'].shift(1)

# (3) 이동 평균 (최근 3일 판매 흐름)
df_daily['수량_MA3'] = df_daily.groupby('상품명')['수량'].transform(lambda x: x.rolling(window=3).mean()).shift(1)

# (4) 증가량 및 변화율
df_daily['주문_증가량'] = df_daily['수량_lag_1'] - df_daily.groupby('상품명')['수량'].shift(2)
df_daily['평점_변화'] = df_daily['평점'] - df_daily['평점_lag_1']

# 결측치 제거 (Lag 생성으로 인해 생긴 빈 칸)
df_model = df_daily.dropna().copy()

# 4. 인코딩 (문자열 -> 숫자)
le_cols = ['상품명', '카테고리', '브랜드']
for col in le_cols:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# 5. 데이터 분할 (8:2 시간순 분할)
features = [
    '상품명', '카테고리', '브랜드', 'month', 'day_of_week', 'is_weekend',
    '수량_lag_1', '수량_lag_7', '수량_MA3', '주문_증가량', '평점_변화', '배송소요시간'
]
target = '수량' # 매출을 예측하려면 '금액'으로 변경

split_idx = int(len(df_model) * 0.8)
train = df_model.iloc[:split_idx]
test = df_model.iloc[split_idx:]

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

# 6. 모델 학습 (RandomForest)
# XGBoost 라이브러리가 설치되어 있다면 xgb.XGBRegressor로 교체 가능합니다.
model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

# 7. 예측 및 성능 평가
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
# 정확도 계산: (1 - 평균오차율)
accuracy = (1 - (mae / y_test.mean())) * 100

print(f"--- 수요 예측 모델 평가 결과 ---")
print(f"1. MAE (평균 오차 수량): {mae:.2f}개")
print(f"2. RMSE (제곱근 오차): {rmse:.2f}")
print(f"3. R2 Score (설명력): {r2:.4f}")
print(f"4. 예측 정확도: {accuracy:.2f}%")

# 8. 변수 중요도 시각화
importances = model.feature_importances_
plt.figure(figsize=(10, 6))
plt.barh(features, importances)
plt.title('Feature Importances for Demand Forecasting')
plt.xlabel('Importance Value')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import platform

# 1. 한글 폰트 및 마이너스 깨짐 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin': # Mac
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 로드
df_final = pd.read_csv('blinkit_final_data.csv')
df_marketing = pd.read_csv(r'data\blinkit_marketing_performance.csv')

# 날짜 형식 변환
df_final['주문날짜'] = pd.to_datetime(df_final['주문날짜'])
df_marketing['date'] = pd.to_datetime(df_marketing['date'])

# 3. 마케팅 데이터 전처리 및 결합 준비
# 회원타입 매핑 (주문 데이터의 회원타입을 마케팅 타겟 오디언스 명칭과 맞춤)
mapping = {
    'Regular': 'All',
    'New': 'New Users',
    'Inactive': 'Inactive',
    'Premium': 'Premium'
}
df_final['target_audience'] = df_final['회원타입'].map(mapping)

# 마케팅 데이터 일별/타겟별 합산 (같은 날 여러 채널 광고 집계)
df_mkt_agg = df_marketing.groupby(['date', 'target_audience']).agg({
    'spend': 'sum',
    'clicks': 'sum',
    'roas': 'mean'
}).reset_index()

# 주문 데이터와 마케팅 데이터 결합
df_combined = pd.merge(
    df_final, 
    df_mkt_agg, 
    left_on=['주문날짜', 'target_audience'], 
    right_on=['date', 'target_audience'], 
    how='left'
).fillna(0) # 마케팅 활동이 없는 날은 0으로 채움

# 4. 분석 대상 상품 리스트
target_products = [
    'Toilet Cleaner', 'Cough Syrup', 'Lotion', 
    'Dish Soap', 'Vitamins', 'Baby Wipes'
]

final_results = []

print("--- 상품별 수요 예측 모델 학습 시작 ---")

for product in target_products:
    # (1) 특정 상품 필터링
    df_p = df_combined[df_combined['상품명'] == product].copy()
    
    # (2) 일별 집계 (수요 예측은 일일 단위가 기본)
    df_daily = df_p.groupby('주문날짜').agg({
        '수량': 'sum',
        '평점': 'mean',
        'spend': 'sum',
        'clicks': 'sum'
    }).reset_index().sort_values('주문날짜')
    
    # (3) 시계열 피처 생성 (Lag 변수)
    df_daily['수량_lag_1'] = df_daily['수량'].shift(1) # 어제 판매량
    df_daily['수량_lag_7'] = df_daily['수량'].shift(7) # 지난주 오늘 판매량
    df_daily['주문_증가량'] = df_daily['수량_lag_1'] - df_daily['수량'].shift(2) # 최근 추세
    
    # (4) 결측치 제거 (Lag 생성으로 생긴 빈 칸)
    df_model = df_daily.dropna()
    
    if len(df_model) > 30: # 최소 데이터 개수 확인
        # 피처 및 타겟 설정
        features = ['수량_lag_1', '수량_lag_7', '주문_증가량', '평점', 'spend', 'clicks']
        X = df_model[features]
        y = df_model['수량']
        
        # 데이터 분할 (시간 순서대로 80% 학습, 20% 테스트)
        split = int(len(X) * 0.8)
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]
        
        # 모델 학습 (Random Forest)
        model = RandomForestRegressor(n_estimators=100, random_state=42)
        model.fit(X_train, y_train)
        
        # 예측 및 평가
        y_pred = model.predict(X_test)
        mae = mean_absolute_error(y_test, y_pred)
        # 정확도 산식: 1 - (평균오차 / 실제평균값)
        accuracy = (1 - (mae / y_test.mean())) * 100
        
        final_results.append({
            '상품명': product,
            '데이터_일수': len(df_model),
            '평균오차(MAE)': f"{mae:.2f}개",
            '정확도': f"{accuracy:.2f}%"
        })
    else:
        final_results.append({
            '상품명': product,
            '데이터_일수': len(df_model),
            '평균오차(MAE)': '데이터 부족',
            '정확도': '학습 불가'
        })

# 5. 최종 결과 리포트 출력
report_df = pd.DataFrame(final_results)
print("\n[최종 수요 예측 성능 리포트]")
print(report_df)

# 6. 결과 시각화
plt.figure(figsize=(12, 6))
# 정확도 수치에서 '%' 제거 후 숫자로 변환하여 시각화
plot_data = report_df[report_df['정확도'] != '학습 불가'].copy()
plot_data['정확도_float'] = plot_data['정확도'].str.replace('%', '').astype(float)

plt.bar(plot_data['상품명'], plot_data['정확도_float'], color='skyblue')
plt.axhline(y=60, color='red', linestyle='--', label='평균 정확도 라인(60%)')
plt.title('상위 6개 상품별 수요 예측 정확도 비교', fontsize=15)
plt.ylabel('정확도 (%)')
plt.ylim(0, 100)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import platform

# 1. 한글 깨짐 방지 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':
    plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 로드
df_final = pd.read_csv('blinkit_final_data.csv')
df_marketing = pd.read_csv(r'data\blinkit_marketing_performance.csv')
df_inv = pd.read_csv(r'data\blinkit_inventory.csv') # 업로드하신 파일명에 맞춰 수정

# 날짜 형식 변환 (중요!)
df_final['주문날짜'] = pd.to_datetime(df_final['주문날짜'])
df_marketing['date'] = pd.to_datetime(df_marketing['date'])
# inventory의 날짜는 '17-03-2023' 형태이므로 dayfirst=True 또는 format 지정
df_inv['date'] = pd.to_datetime(df_inv['date'], dayfirst=True)

# 3. 마케팅 데이터 결합 (타겟 고객 매핑 포함)
mkt_mapping = {'Regular': 'All', 'New': 'New Users', 'Inactive': 'Inactive', 'Premium': 'Premium'}
df_final['target_audience'] = df_final['회원타입'].map(mkt_mapping)

df_mkt_agg = df_marketing.groupby(['date', 'target_audience']).agg({
    'spend': 'sum', 'clicks': 'sum', 'roas': 'mean'
}).reset_index()

# 4. 재고 데이터 결합 (주문날짜 + 상품_ID 기준)
# 재고 데이터와 주문 데이터를 합쳐서 "물건이 있었는지"를 모델이 알게 합니다.
df_master = pd.merge(df_final, df_mkt_agg, 
                     left_on=['주문날짜', 'target_audience'], 
                     right_on=['date', 'target_audience'], how='left')

df_master = pd.merge(df_master, df_inv, 
                     left_on=['주문날짜', '상품_ID'], 
                     right_on=['date', 'product_id'], how='left').fillna(0)

# 5. 시계열 모델링을 위한 일별/상품별 집계
df_daily = df_master.groupby(['주문날짜', '상품명']).agg({
    '수량': 'sum',
    '평점': 'mean',
    'spend': 'mean',
    'clicks': 'mean',
    'stock_received': 'sum',  # 재고 입고량
    'damaged_stock': 'sum',   # 재고 파손량
    '카테고리': 'first',
    '브랜드': 'first'
}).reset_index().sort_values(['상품명', '주문날짜'])

# 6. 핵심 피처 생성 (Lag 변수)
df_daily['수량_lag_1'] = df_daily.groupby('상품명')['수량'].shift(1)
df_daily['수량_lag_7'] = df_daily.groupby('상품명')['수량'].shift(7)
df_daily['입고_lag_1'] = df_daily.groupby('상품명')['stock_received'].shift(1)

df_model = df_daily.dropna().copy()

# 7. 범주형 변수 처리 (Label Encoding)
le = LabelEncoder()
for col in ['상품명', '카테고리', '브랜드']:
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# 8. 모델 학습 및 성과 측정
features = ['상품명', '카테고리', '브랜드', '수량_lag_1', '수량_lag_7', 
            'spend', 'clicks', 'stock_received', 'damaged_stock', '입고_lag_1']
X = df_model[features]
y = df_model['수량']

split = int(len(X) * 0.8)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X[:split], y[:split])

# 결과 출력
y_pred = model.predict(X[split:])
mae = mean_absolute_error(y[split:], y_pred)
accuracy = (1 - (mae / y[split:].mean())) * 100

print(f"--- 최종 통합 모델 성과 ---")
print(f"평균 오차(MAE): {mae:.2f}개")
print(f"예측 정확도: {accuracy:.2f}%")

# 9. 변수 중요도 시각화

plt.figure(figsize=(10, 6))
importances = model.feature_importances_
plt.barh(features, importances, color='teal')
plt.title('수요 예측 변수 중요도 (주문+마케팅+재고)')
plt.xlabel('중요도')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# 1. 환경 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 로드 및 기본 전처리
# (이미 통합된 df_master 데이터가 있다고 가정합니다)
# df_master = [주문 + 마케팅 + 재고 통합본]

def train_and_evaluate(data, target_col='수량'):
    # 피처 생성 (Lag 1, Lag 2)
    data['lag_1'] = data[target_col].shift(1)
    data['lag_2'] = data[target_col].shift(2)
    df_model = data.dropna()
    
    X = df_model[['lag_1', 'lag_2', 'spend', 'clicks']]
    y = df_model[target_col]
    
    split = int(len(X) * 0.8)
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X[:split], y[:split])
    
    y_pred = model.predict(X[split:])
    mae = mean_absolute_error(y[split:], y_pred)
    accuracy = (1 - (mae / y[split:].mean())) * 100
    return accuracy

# --- [분석 대상: Pet Treats] ---
target_product = 'Pet Treats'
df_target = df_master[df_master['상품명'] == target_product].copy()

# A. 일별 데이터 정확도 측정
df_daily = df_target.groupby('주문날짜').agg({
    '수량': 'sum', 'spend': 'sum', 'clicks': 'sum'
}).reset_index().sort_values('주문날짜')
daily_acc = train_and_evaluate(df_daily)

# B. 주별 데이터 정확도 측정 (Resampling)
df_weekly = df_target.set_index('주문날짜').resample('W').agg({
    '수량': 'sum', 'spend': 'sum', 'clicks': 'sum'
}).reset_index()
weekly_acc = train_and_evaluate(df_weekly)

# 3. 정확도 비교 시각화
labels = ['일별 모델 (Daily)', '주별 모델 (Weekly)']
accuracies = [daily_acc, weekly_acc]

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, accuracies, color=['lightgray', 'royalblue'], width=0.6)

# 그래프 위에 숫자 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.2f}%', ha='center', va='bottom', fontweight='bold')

plt.title(f'[{target_product}] 데이터 단위별 예측 정확도 비교', fontsize=14)
plt.ylabel('정확도 (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f"일별 정확도: {daily_acc:.2f}%")
print(f"주별 정확도: {weekly_acc:.2f}%")
print(f"정확도 상승분: {weekly_acc - daily_acc:.2f}%p")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 모든 데이터 로드 및 날짜 전처리
df_sales = pd.read_csv('blinkit_final_data.csv')
df_marketing = pd.read_csv(r'data\blinkit_marketing_performance.csv')
df_inventory = pd.read_csv(r'data\blinkit_inventory.csv')

df_sales['주문날짜'] = pd.to_datetime(df_sales['주문날짜'])
df_marketing['date'] = pd.to_datetime(df_marketing['date'])
df_inventory['date'] = pd.to_datetime(df_inventory['date'], dayfirst=True)

# 2. 데이터 통합 (Merge)
mkt_map = {'Regular': 'All', 'New': 'New Users', 'Inactive': 'Inactive', 'Premium': 'Premium'}
df_sales['target_audience'] = df_sales['회원타입'].map(mkt_map)

df_mkt_agg = df_marketing.groupby(['date', 'target_audience']).agg({
    'spend': 'sum', 'clicks': 'sum'
}).reset_index()

# 3단계 통합: 판매 + 마케팅 + 재고
df_master = pd.merge(df_sales, df_mkt_agg, left_on=['주문날짜', 'target_audience'], right_on=['date', 'target_audience'], how='left')
df_master = pd.merge(df_master, df_inventory, left_on=['주문날짜', '상품_ID'], right_on=['date', 'product_id'], how='left').fillna(0)

# 3. 주별(Weekly) 데이터로 재집계
# '주문날짜'를 기준으로 일요일(W) 단위로 모든 수치를 합산합니다.
df_weekly = df_master.set_index('주문날짜').resample('W').agg({
    '수량': 'sum',
    '금액': 'sum',
    '평점': 'mean',
    'spend': 'sum',         # 주간 총 광고비
    'clicks': 'sum',        # 주간 총 클릭수
    'stock_received': 'sum', # 주간 총 입고량
    'damaged_stock': 'sum'   # 주간 총 파손량
}).reset_index()

# 4. 주별 상관관계 분석 및 시각화
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
plt.rcParams['axes.unicode_minus'] = False

corr_cols = ['수량', '금액', '평점', 'spend', 'clicks', 'stock_received', 'damaged_stock']
weekly_corr = df_weekly[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(weekly_corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('주별(Weekly) 판매-마케팅-재고 상관관계 분석')
plt.show()

print("--- 주간 데이터 샘플 ---")
print(df_weekly.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 분석 대상 Top 10 상품 리스트 설정
top_10_products = [
    'Pet Treats', 'Toilet Cleaner', 'Lotion', 'Vitamins', 'Dish Soap',
    'Baby Wipes', 'Cough Syrup', 'Cat Food', 'Pulses', 'Orange Juice'
]

# 2. 주별+제품별 데이터에서 Top 10만 추출
df_top_10 = df_weekly_product[df_weekly_product['상품명'].isin(top_10_products)]

# 3. 상품별 상관관계 비교 분석
# 각 상품별로 '수량'과의 상관계수만 모아서 한눈에 비교합니다.
corr_summary = []

for product in top_10_products:
    product_df = df_top_10[df_top_10['상품명'] == product]
    if len(product_df) > 1:
        # 상관계수 행렬에서 '수량' 컬럼만 가져옴
        product_corr = product_df[['수량', '금액', '평점', 'spend', 'clicks', 'stock_received', 'damaged_stock']].corr()['수량']
        product_corr.name = product
        corr_summary.append(product_corr)

# 결과 합치기
df_corr_comparison = pd.concat(corr_summary, axis=1).T.drop(columns=['수량'])

# 4. 시각화: 상품별 마케팅/재고 민감도 히트맵
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr_comparison, annot=True, cmap='RdYlGn', center=0, fmt='.2f')
plt.title('Top 10 상품별 주요 지표 상관관계 비교 (수량 기준)', fontsize=15)
plt.ylabel('상품명')
plt.xlabel('영향 지표')
plt.tight_layout()
plt.show()

# 5. 분석 결과 CSV 저장 (팀 보고서용)
df_corr_comparison.to_csv('top_10_correlation_analysis.csv', encoding='utf-8-sig')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# 1. 데이터 준비 (Top 10 상품 대상)
top_10 = ['Pet Treats', 'Toilet Cleaner', 'Lotion', 'Vitamins', 'Dish Soap', 
          'Baby Wipes', 'Cough Syrup', 'Cat Food', 'Pulses', 'Orange Juice']
df_model_data = df_weekly_product[df_weekly_product['상품명'].isin(top_10)].copy()

# 2. 변수 정규화 (Min-Max Scaling)
# 이유: 금액(수만 단위)과 평점(1~5 단위)은 스케일이 다르므로 0~1 사이로 맞춰야 가중치 계산이 정확해집니다.
scaler = MinMaxScaler()
features_to_scale = ['금액', '평점', 'spend', 'clicks', 'stock_received', 'damaged_stock']
df_model_data[features_to_scale] = scaler.fit_transform(df_model_data[features_to_scale])

# 3. 가중합 피처(Weighted Feature) 생성
# df_corr_comparison에 저장된 상관계수 값을 가중치로 사용합니다.
# 예: 수량 = (평점 * 평점_상관계수) + (광고비 * 광고비_상관계수) ...
def apply_weights(row):
    product = row['상품명']
    if product in df_corr_comparison.index:
        weights = df_corr_comparison.loc[product]
        # 각 피처에 상관계수를 곱한 값을 합산하여 '예측 지수'를 만듭니다.
        weighted_score = (
            row['평점'] * weights['평점'] +
            row['spend'] * weights['spend'] +
            row['clicks'] * weights['clicks'] +
            row['stock_received'] * weights['stock_received'] +
            row['damaged_stock'] * weights['damaged_stock']
        )
        return weighted_score
    return 0

df_model_data['weighted_score'] = df_model_data.apply(apply_weights, axis=1)

# 4. 시계열 특성 추가 (전주 판매량 - Lag Feature)
# 모델이 과거의 흐름을 알 수 있게 전주 수량을 추가합니다.
df_model_data['lag_1'] = df_model_data.groupby('상품명')['수량'].shift(1)
df_model_data = df_model_data.dropna() # 결측치 제거

# 5. 모델링 및 예측 (Random Forest)
def train_and_predict(product_name):
    product_df = df_model_data[df_model_data['상품명'] == product_name]
    
    # 학습 변수: 가중합 점수 + 전주 수량 + 개별 지표들
    X = product_df[['weighted_score', 'lag_1', 'spend', '평점', 'stock_received']]
    y = product_df['수량']
    
    # 데이터 분할 (과거 80%로 학습, 최근 20%로 테스트)
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    # 성능 지표 계산
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return y_test, y_pred, mae, r2

# 6. 결과 확인 (예: Pet Treats)
y_test, y_pred, mae, r2 = train_and_predict('Pet Treats')

print(f"--- Pet Treats 예측 결과 ---")
print(f"평균 절대 오차(MAE): {mae:.2f}")
print(f"결정계수(R2 Score): {r2:.2f}")

# 시각화
plt.figure(figsize=(10, 5))
plt.plot(y_test.values, label='실제 판매량', marker='o')
plt.plot(y_pred, label='예측 판매량', marker='x')
plt.title('Pet Treats 주간 수요 예측 결과 (가중치 모델)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# 1. 데이터 준비 및 상품 필터링
top_10 = ['Pet Treats', 'Toilet Cleaner', 'Lotion', 'Vitamins', 'Dish Soap', 
          'Baby Wipes', 'Cough Syrup', 'Cat Food', 'Pulses', 'Orange Juice']
df_model_data = df_weekly_product[df_weekly_product['상품명'].isin(top_10)].copy()

# 2. 변수 정규화 (Min-Max Scaling)
# 분석 전 수치들의 단위를 0~1 사이로 맞춥니다.
scaler = MinMaxScaler()
features_to_scale = ['금액', '평점', 'spend', 'clicks', 'stock_received', 'damaged_stock']
df_model_data[features_to_scale] = scaler.fit_transform(df_model_data[features_to_scale])

# 3. 고도화된 가중합 피처(Weighted Feature) 생성
# 상관계수의 제곱을 가중치로 사용하여 중요한 변수의 영향력을 증폭시킵니다.
def apply_optimized_weights(row):
    product = row['상품명']
    if product in df_corr_comparison.index:
        weights = df_corr_comparison.loc[product]
        # 상관계수가 높은 피처의 비중을 비선형적으로 높임 (제곱 활용)
        weighted_score = (
            row['평점'] * (weights['평점']**2) +
            row['spend'] * (weights['spend']**2) +
            row['clicks'] * (weights['clicks']**2) +
            row['stock_received'] * (weights['stock_received']**2)
        )
        return weighted_score
    return 0

df_model_data['weighted_score'] = df_model_data.apply(apply_optimized_weights, axis=1)

# 4. 상급 시계열 피처 생성 (데이터 보강)
# 과거 1~2주 데이터와 최근 추세를 모델에 학습시킵니다.
df_model_data['lag_1'] = df_model_data.groupby('상품명')['수량'].shift(1)
df_model_data['lag_2'] = df_model_data.groupby('상품명')['수량'].shift(2)
# 3주 이동 평균: 최근 판매 레벨의 기준점 제공
df_model_data['rolling_mean_3'] = df_model_data.groupby('상품명')['수량'].transform(lambda x: x.rolling(window=3).mean())
# 모멘텀: 판매 가속도 (늘고 있는가 줄고 있는가)
df_model_data['momentum'] = df_model_data['lag_1'] - df_model_data['lag_2']

# 결측치 제거 (Lag 생성으로 인해 생긴 초기 데이터 제외)
df_model_data = df_model_data.dropna()

# 5. 모델링 및 예측 함수 정의
def train_and_predict_v3(product_name):
    product_df = df_model_data[df_model_data['상품명'] == product_name]
    
    # 학습 변수 확장: 가중합 점수 + 시계열 피처들 + 원본 마케팅/평점 지표
    X = product_df[['weighted_score', 'lag_1', 'lag_2', 'rolling_mean_3', 'momentum', 'spend', '평점']]
    y = product_df['수량']
    
    # 시계열 특성을 고려하여 최근 데이터를 테스트셋으로 분리 (8:2)
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    
    # 모델 설정: 나무 개수를 늘려 안정성 확보 및 과적합 방지(max_depth)
    model = RandomForestRegressor(n_estimators=300, max_depth=7, random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    # 성능 지표 계산
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    # 정확도: 1 - (평균 오차 / 실제 평균 판매량)
    accuracy = (1 - (mae / y_test.mean())) * 100 if y_test.mean() > 0 else 0
    
    return y_test, y_pred, mae, r2, accuracy

# 6. 전체 Top 10 상품 성능 비교 실행
performance_results = []

for product in top_10:
    try:
        y_test, y_pred, mae, r2, acc = train_and_predict_v3(product)
        performance_results.append({
            '상품명': product,
            'MAE(오차)': round(mae, 2),
            'R2 Score': round(r2, 2),
            '정확도(%)': round(acc, 2)
        })
    except Exception as e:
        print(f"{product} 분석 중 오류: {e}")

# 결과 출력 및 데이터프레임 정리
df_perf_final = pd.DataFrame(performance_results).sort_values(by='정확도(%)', ascending=False)
print("--- [고도화 모델] Top 10 전 상품 예측 성능 ---")
print(df_perf_final)

# 7. 시각화 (정확도 비교)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(12, 6))
bars = plt.bar(df_perf_final['상품명'], df_perf_final['정확도(%)'], color='#2E86C1')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1, f'{height}%', ha='center', va='bottom', fontweight='bold')

plt.title('가중치 고도화 및 시계열 피처 추가 후 상품별 정확도', fontsize=15)
plt.ylabel('정확도 (%)')
plt.ylim(0, 110)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, r2_score

# 1. 데이터 준비 (앞서 만든 df_model_data 활용)
# ARIMA는 시간 순서가 중요하므로 인덱스를 날짜로 설정하고 정렬합니다.
def train_and_predict_arima(product_name):
    # 특정 상품 데이터 필터링
    product_df = df_model_data[df_model_data['상품명'] == product_name].sort_values('주문날짜')
    
    # 목표 변수(y)와 외부 변수(exog) 설정
    # ARIMA는 스스로 과거를 보므로 lag 변수는 제외하고 외부 충격량인 weighted_score와 spend만 넣습니다.
    y = product_df['수량']
    exog = product_df[['weighted_score', 'spend', '평점']]
    
    # 데이터 분할 (8:2)
    split = int(len(y) * 0.8)
    y_train, y_test = y[:split], y[split:]
    exog_train, exog_test = exog[:split], exog[split:]
    
    # 2. SARIMAX 모델 설정 및 학습
    # order(p,d,q): p(자기회귀), d(차분), q(이동평균)
    # 주별 데이터이므로 적절한 기본값 (1, 1, 1)을 설정합니다.
    model = SARIMAX(y_train, 
                    exog=exog_train, 
                    order=(1, 1, 1), 
                    enforce_stationarity=False, 
                    enforce_invertibility=False)
    
    model_fit = model.fit(disp=False)
    
    # 3. 예측 실행
    # 테스트 기간만큼의 외부 변수(exog_test)를 넣어 미래를 예측합니다.
    forecast = model_fit.get_forecast(steps=len(y_test), exog=exog_test)
    y_pred = forecast.predicted_mean
    
    # 4. 성능 지표 계산
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    accuracy = (1 - (mae / y_test.mean())) * 100 if y_test.mean() > 0 else 0
    
    return y_test, y_pred, mae, r2, accuracy

# 5. Top 10 상품에 대해 ARIMA 실행 및 결과 수집
arima_results = []
for product in top_10:
    try:
        y_test, y_pred, mae, r2, acc = train_and_predict_arima(product)
        arima_results.append({
            '상품명': product,
            'ARIMA_MAE': round(mae, 2),
            'ARIMA_R2': round(r2, 2),
            'ARIMA_정확도(%)': round(acc, 2)
        })
    except:
        continue

df_arima_perf = pd.DataFrame(arima_results).sort_values(by='ARIMA_정확도(%)', ascending=False)
print("--- [ARIMA 모델] 예측 성능 결과 ---")
print(df_arima_perf)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

def train_and_predict_ensemble(product_name):
    # 1. 개별 모델의 예측값 가져오기
    # 앞에서 만든 train_and_predict_v3(RF)와 train_and_predict_arima 함수를 호출합니다.
    y_test_rf, y_pred_rf, mae_rf, r2_rf, acc_rf = train_and_predict_v3(product_name)
    y_test_ari, y_pred_ari, mae_ari, r2_ari, acc_ari = train_and_predict_arima(product_name)
    
    # 2. 앙상블 가중치 설정 (성능 기반 자동 조절)
    # R2 Score가 더 높은 모델에 더 많은 비중을 둡니다. 
    # 만약 특정 모델의 R2가 너무 낮으면(음수 등) 최소 비중을 보장합니다.
    w_rf = max(r2_rf, 0.1)
    w_ari = max(r2_ari, 0.1)
    
    # 가중치 정규화 (합이 1이 되도록)
    total_w = w_rf + w_ari
    weight_rf = w_rf / total_w
    weight_ari = w_ari / total_w
    
    # 3. 최종 예측치 계산 (Weighted Sum)
    y_pred_ensemble = (y_pred_rf * weight_rf) + (y_pred_ari * weight_ari)
    
    # 4. 앙상블 성능 지표 계산
    mae_ens = mean_absolute_error(y_test_rf, y_pred_ensemble)
    r2_ens = r2_score(y_test_rf, y_pred_ensemble)
    accuracy_ens = (1 - (mae_ens / y_test_rf.mean())) * 100
    
    return y_test_rf, y_pred_ensemble, mae_ens, r2_ens, accuracy_ens, weight_rf

# --- Top 10 전체 앙상블 결과 확인 ---
ensemble_results = []
for product in top_10:
    try:
        y_test, y_pred, mae, r2, acc, w_rf = train_and_predict_ensemble(product)
        ensemble_results.append({
            '상품명': product,
            'Ensemble_MAE': round(mae, 2),
            'Ensemble_R2': round(r2, 2),
            '최종_정확도(%)': round(acc, 2),
            'RF비중': round(w_rf, 2)
        })
    except:
        continue

df_ensemble_perf = pd.DataFrame(ensemble_results).sort_values(by='최종_정확도(%)', ascending=False)
print("--- [Ensemble 모델] 최종 예측 성능 ---")
print(df_ensemble_perf)
```
